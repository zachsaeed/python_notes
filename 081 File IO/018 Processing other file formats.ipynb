csv, json and xml files are basically text files formatted in a specific way and have separate extensions. 

Therefore, can use python's builtin open() method to to get an extIOWrapper. This object can be passed on to specific read/ write functions which are programmed to read data as per the respective schema. We will discuss them below:

##  CSV module
Python comes built-in with a module called csv which makes it easy for us to process data in csv files.

```python
import csv

with open('./files/cities.csv') as csv_file:
    print(csv_file)
    
    obj_csv_reader = csv.reader(csv_file, delimiter=',')
    print(obj_csv_reader)
    
    
    line_count = 0
    for row in obj_csv_reader:
        #print('PRINT ROW: ')
        #print(type(row))
        #print(row)
        
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        else:
            print(f'\t The city, {row[8]} is in {row[9]} state.')
            line_count += 1
            
    print(f'Processed {line_count} lines.')
    
```
